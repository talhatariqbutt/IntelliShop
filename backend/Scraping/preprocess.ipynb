```python
# Remove duplicates from df2 based on 'Downloaded_Image_Id'
df1_unique = df1.drop_duplicates(subset='Downloaded Image ID')
```

```python
df1_unique
```

```python
# Merge only the 'Url' column from df1 to df2 based on 'Downloaded_Image_Id'
merged_df = pd.merge(df2, df1_unique[['Downloaded Image ID', 'Downloaded Image URL']], left_on='image_path', right_on='Downloaded Image ID', how='left')
merged_df.drop('Downloaded Image ID', axis=1, inplace=True)
merged_df
```

```python
merged_df.to_csv('image_index_urls.csv', index=False)
```

```python
# Modify the columns you want to include from df2
columns_to_merge = ['Downloaded Image ID', 'Downloaded Image URL', 'Product Name', 'Price', 'Product ID']

# Perform the merge
merged_df = pd.merge(df2, df1_unique[columns_to_merge], left_on='image_path', right_on='Downloaded Image ID', how='left')

# # Drop the 'Downloaded_Image_Id' column if you don't need it anymore
merged_df.drop('Downloaded Image ID', axis=1, inplace=True)

# Display the result
merged_df
```

```python
# rename column
merged_df.rename(columns={'image_path': 'Downloaded Image Name'}, inplace=True)
merged_df
```

```python
# change the order of columns
merged_df = merged_df[['Product ID','Product Name', 'Price', 'Downloaded Image Name', 'Downloaded Image URL']]
merged_df
```

```python
merged_df.to_csv('image_index_urls.csv', index=False)
```
